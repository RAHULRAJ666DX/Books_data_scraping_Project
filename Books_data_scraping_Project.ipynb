### **Importing Libraries**

---



```python
import requests
import pandas as pd
```

```python
# Beautiful Soup Library

from bs4 import BeautifulSoup
```

**Website URL**

```python
url = "https://books.toscrape.com/"
```

**Using the requests.get we are fetching the data from url** 

```python
response = requests.get(url)
```

```python
response
```

Response [200] means the task has been completed

```python
response = response.content
response
```

### **We are getting the data but the data is not in a readable format**

**Creating a soup object to parse html data**

```python
soup = BeautifulSoup(response,'html.parser')
```

```python
soup
```

**Using Prettify to improve data with proper indents and easy reable format**

```python
print(soup.prettify())
```

## **Since all the books are inside the Ordered List we need to fetch the OL first into a variable to iterate through it**

```python
ol = soup.find('ol')
```

```python
ol
```

### To find all the instances of the ol articles 

## **We are going to shortlist the entire content into article using class name**  
### **To specify the articles of one particular CLASS we will use find_all**

```python
articles = ol.find_all('article', class_='product_pod')
```

```python
articles
```

# **Now to fetch all the articles image alt value , the star class and the price**
## **Creating a for loop**

### we are grabbing the image title which is in a alt attribute of **Image** class   


```python
# we are grabbing the image title which is in a alt attribute under image  
for article in articles:
  image = article.find('img')
  title = image.attrs['alt']
  print(title)
```

We fetched the Title from 20 books from the webpage 

### **Getting the Rating of all the Books under P class**

```python
for article in articles:
  # rating = article.find('img')
  rating = article.find('p')
  rating = rating['class']
  print(rating[1])
```

### **Getting the Price of all the books on the webpage which is in p class using class name**

```python
for article in articles:
  # rating = article.find('img')
  # rating = article.find('p')
  # rating = rating['class']
  price = article.find('p',class_='price_color')
  print(price)  
```

**But its printing the entire pclass to get only the Text form price of the P class** 

we can use .text with find operation
and iterate for price[1:] with float data type 

```python
for article in articles:
  # rating = article.find('img')
  # rating = article.find('p')
  # rating = rating['class']
  price = article.find('p',class_='price_color').text
  price = float(price[1:])
  print(price)  
```

### **Getting the stock availability of all the books on the webpage which is in p class using class name**

```python
for article in articles:
  # rating = article.find('img')
  # rating = article.find('p')
  # rating = rating['class']
  # price = article.find('p',class_='price_color')
  stock = article.find('p',class_='instock availability').text.strip()
  print(stock)  
```

## **Now Combining all the variables**

We have a list of lists of Books Title, Price and its Rating
and we have a total of 20 books information from page 1

```python
books = []
for article in articles:
  image = article.find('img')
  title = image.attrs['alt']
  
  rating1 = article.find('p')
  rating = rating1['class'][1]
  
  price = article.find('p',class_='price_color').text
  price = float(price[1:])
  
  stock = article.find('p',class_='instock availability').text.strip()

  books.append([title,price,rating,stock])
print(books)
print(len(books))
```

### **We successfully fetched data for "title","price","rating" and "stock" from webpage**

## If we need data collected from all the **50 pages** of a website we have to create a for loop with **range(1,51)** for the URL link to iterate from page **1 to 50**

### we have to include all the previous commands inside a single code snippet

### **Fetching the Data from the first 5 pages to verify**

```python
book1 = []
for i in range(1,6):
  url = f"https://books.toscrape.com/catalogue/page-{i}.html"
  response = requests.get(url)
  response = response.content
  soup = BeautifulSoup(response, 'html.parser')
  ol = soup.find('ol')
  articles = ol.find_all('article', class_='product_pod')
  for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    rating1 = article.find('p')
    rating = rating1['class'][1]
    price = article.find('p', class_='price_color').text
    price = float(price[1:])
    stock = article.find('p',class_='instock availability').text.strip()

    book1.append([title, rating, price, stock])
print(len(book1))
print(book1)
```

**We have data from 5 webpages and we will convert them to DataFrame using Pandas library**

```python
df1 = pd.DataFrame(book1,columns=['Book Name','Ratings','Price','Stock'])
```

```python
df1.head()
```

Converting to **CSV**

```python
df1.to_csv('Book1.csv')
```

# **Now that we know we can successfully fetch and convert the data from the first 5 Pages we can do the same for all 50 pages by using range(1,51)**

```python
book_full50 = []
for i in range(1,51):
  url = f"https://books.toscrape.com/catalogue/page-{i}.html"
  response = requests.get(url)
  response = response.content
  soup = BeautifulSoup(response, 'html.parser')
  ol = soup.find('ol')
  articles = ol.find_all('article', class_='product_pod')
  for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    rating1 = article.find('p')
    rating = rating1['class'][1]
    price = article.find('p', class_='price_color').text
    price = float(price[1:])
    stock = article.find('p',class_='instock availability').text.strip()

    book_full50.append([title, rating, price, stock])
len(book_full50)
```

```python
print("We have a total of {} books".format(len(book_full50)))
```

### **Converting to a DataFrame**

```python
df_50 = pd.DataFrame(book_full50,columns=['Book Name','Ratings','Price','Stock'])
```
