## **If we need data collected from all the 50 pages of a website we have to create a for loop with range(1,51) for the URL link to iterate from page 1 to 50**

we have to include all the previous commands inside a single code snippet

### **Fetching the Data from the first 5 pages to verify**

```python
book1 = []
for i in range(1,6):
  url = f"https://books.toscrape.com/catalogue/page-{i}.html"
  response = requests.get(url)
  response = response.content
  soup = BeautifulSoup(response, 'html.parser')
  ol = soup.find('ol')
  articles = ol.find_all('article', class_='product_pod')
  for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    rating1 = article.find('p')
    rating = rating1['class'][1]
    price = article.find('p', class_='price_color').text
    price = float(price[1:])
    book1.append([title, rating, price])
```

We have data from 5 webpages and we will convert them to DataFrame using Pandas library  

```python
df1 = pd.DataFrame(book1,columns=['Book Name','Ratings','Price'])
```

Now that we know we can successfully fetch and convert the data from the first 5 Pages we can do the same for all 50 pages by using range(1,51)

```python
book_full50 = []
for i in range(1,51):
  url = f"https://books.toscrape.com/catalogue/page-{i}.html"
  response = requests.get(url)
  response = response.content
  soup = BeautifulSoup(response, 'html.parser')
  ol = soup.find('ol')
  articles = ol.find_all('article', class_='product_pod')
  for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    rating1 = article.find('p')
    rating = rating1['class'][1]
    price = article.find('p', class_='price_color').text
    price = float(price[1:])
    book_full50.append([title, rating, price])
```

```python
df_50 = pd.DataFrame(book_full50,columns=['Book Name','Ratings','Price'])
```
